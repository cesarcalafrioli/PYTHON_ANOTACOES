## **Literals**

Literal é uma notação para representar alguns valores fixos no código.É um dado cujos valores são determinados pela literal em sí, ou seja, 123 é 123, já o caractere c pode ser qualquer coisa, sendo assim não literal. É utilizado para codificar dados e colocá-los no código.



```python
#Aqui temos dois tipos diferentes de literais. O primeiro é um string, enquanto
# o segundo é um número inteiro.
#
# A função print() apresenta os dois exatamente da mesma forma. No entanto, na
#memória do computador, os dois valores são armazenados de maneira distinta,
#onde, enquanto o string existe apenas como strint, um conjunto de caracteres,
#o inteiro é convertido na representação da máquina (conjunto de bits).
print("2") # A função print mostra ambas as literais na forma legível para humanos
print(2)
```

### **Inteiros**

Inteiro é um tipo de dados desprovido de partes fracionárias, podendo ser tanto negativo quanto positivo. A linguagem permite escrever um inteiro no formato octal, hexadecimal e literais binários.

```python
print(11111111)
print(11_111_111) # O recurso de underscore em números inteiros foi introduzido no Python 3.6
```

Se um valor interiro for precedido por um prefixo 0o ou 0O, será tratado como um valor octal. Dessa forma, o intervalo limite será entre 0 e 7.

```python
print(0o123)
print(0O123)
```

Se o valor inteiro for precediro pelo prefixo 0x ou 0X (zero-x), será então hexadecimal.

```python
print(0xFF)
print(0x123)
```

### **Float**

Valores float possuem (ou podem possuir) uma parte fracionária após o ponto decimal. Este tipo de dados foi projetado para representar e armazenar números que posuam valores fracionários não-vazios.



```python
print(2.5)
print(-0.4)
```

É possível escrever um valor decimal da forma tradicional ou, simplesmente, omitir o zero antes ou depois do ponto decimal, como no exemplo abaixo:

```python
print(0.4)
print(.4)
print(4.0)
print(4.)
```

Não é apenas o ponto decimal que transforma um número em float. Para números maiores é possível utilizar o caracterere "e", tanto maiúsculo quanto minúsculo. Tal caractere é utilizado em notações científicas envolvendo números muito grande ou muito pequeno.

```python
print(2e5)
print(2E5)
print(2.0E5)
print(6.62607E-34)
print(0.0000000000000000000001) # A linguagem Python sempre escolhe a forma mais econômica de apresentar os números. É necessário levar isso em consideração ao criar literais
```

### **Strings**

Utilizado para processar texto, não números, precisa de aspas da mesma forma que os valores float precisam de pontos.

```python
print("I am a string.")

# Se precisa colocar aspas duplas dentro de uma string já delimitado por aspas duplas
print("I like \"Monty Python\"")

# Como alternativa, utilizar apóstrofos em vez de aspas, para inserir palavras ou caracteres com aspas dentro da string.
print('I like "Monty Python"')

# Para o inverso, ou seja, apóstrofo dentro de aspas
print("I'm Monty Python")
print("I\'m Monty Python")
```

### **Boolean**

O valor booleano é utilizado para representar a veracidade de um valor, como por exemplo se um número é maior que outro.

Os valores booleanos possuem apenas dois valores distintos: True ou False, denotado também como 1 e 0.

```python
print(True > False)
print(True < False)
```

### **None**

None é um tipo de dados objeto NoneType utilizado para representar a ausência de um valor.

## **Operadores aritméticos**

Utiliza-se operadores para operar os valores.

```python
# Exponenciação
print(2 ** 3)
print(2 ** 3.)
print(2. ** 3) # Se um dos argumentos for float, o resultado será float.
print(2. ** 3.)

# Multiplicação
print(2 * 3)
print(2 * 3.) # Se um dos argumentos for float, o resultado será float.
print(2. * 3) # Se um dos argumentos for float, o resultado será float.
print(2. * 3.)

# Divisão
print(6 / 3) # Mesmo os dois números sendo inteiros, o resultado será float.
print(6 / 3.)
print(6. / 3)
print(6. / 3.)

# Divisão de inteiros
# A divisão inteira também pode ser chamada de floor division.
print(6 // 3) # Se ambos os números forem inteiros e o resultado não for fracionado, será inteiro também (Arreondado para o menor inteiro)
print(6 // 3.) # Se um dos argumentos for float, o resultado será float.
print(6. // 3)
print(6. // 3.)
print(-6 // 4)
print(6. // -4)

# Resto
"""
Realizar uma divisão por zero;
realizar uma divisão inteira por zero;
encontrar o resto de uma divisão por zero.
"""
print(14 % 4)
print(12 % 4.5)

# Adição
print(-4 + 4)
print(-4. + 8)
print(-4 - 4)
print(4. - 8)
print(-1.1)
print(+2)

# Ordem dos calculos
print(2 ** 2 ** 3) # Neste caso, right-sided binding.
print((5 * ((25 % 13) + 100) / (2 * 13)) // 2) # Subespressões em parênteses são sempre calculados primeiro
print(2 * 3 % 5)
```
