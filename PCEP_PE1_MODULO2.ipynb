### Literals

Literal é um dado cujos valores são determinados pela literal em sí, ou seja, 123 é 123, já o caractere c pode ser qualquer coisa, sendo assim não literal. É utilizado para codificar dados e colocá-los no código.



```python
#Aqui temos dois tipos diferentes de literais. O primeiro é um string, enquanto
# o segundo é um número inteiro.
#
# A função print() apresenta os dois exatamente da mesma forma. No entanto, na
#memória do computador, os dois valores são armazenados de maneira distinta,
#onde, enquanto o string existe apenas como strint, um conjunto de caracteres,
#o inteiro é convertido na representação da máquina (conjunto de bits).
print("2") # A função print mostra ambas as literais na forma legível para humanos
print(2)
```
